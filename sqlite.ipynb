# データの永続化：SQLite

- ソースコード

```python
path = ''
db_name ='car.db'
conn = sqlite3.connect(path + db_name)
cur = conn.cursor()

sql = 'CREATE TABLE cars (id INT , name TEXT, price REAL)'
cur.execute(sql)

conn.close()


```

```python
path = ''
db_name ='car.db'

try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()

    sql = "INSERT INTO cars (id ,name , price )VALUES (1, 'Audi', 52642.00)"
    cur.execute(sql)
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()

    sql = 'CREATE TABLE cars (id INT , name TEXT, price REAL)'
    cur.execute(sql)
except sqlite3.Error as e:
    print(f"An error occurred: {e}")

conn.close()
```
