### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください


```python
s1 = "ヘルシンキ"
s2 = "エストニア"

# zip()で文字をペアにし、リスト内包表記で順番に連結
interleaved_list = [char for pair in zip(s1, s2) for char in pair]

# リストを結合して最終的な文字列を作成
result = "".join(interleaved_list)

print(result)
# 出力: ヘエルスシトンニキア
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
sentence = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"

# 1. split()で単語に分解する
words = sentence.split()

# 2. リスト内包表記で各単語の文字数を取得する
length_list = [len(word) for word in words]

print(length_list)
# 出力: [9, 9, 5, 2, 3, 11, 6, 2, 9, 3, 3, 12]
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
sentence = "It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."

# 対象となる単語の位置 (1-ベース)
target_indices_1based = [1, 2, 5, 6, 8, 9, 16]

# 1. 句読点を削除し、空白で単語に分解する
# 句読点をスペースに置換してからsplit()するのが簡潔です
cleaned_sentence = sentence.replace(',', '').replace('.', '')
words = cleaned_sentence.split()

# 2. 処理結果を格納するための辞書を初期化
result_dict = {}

# 3. 辞書を作成
for i in target_indices_1based:
    # 1-ベースのインデックスを0-ベースに変換
    index_0based = i - 1
    
    # 単語を取得
    word = words[index_0based]
    
    # 単語の先頭2文字を取得 (キー)
    prefix = word[:2]
    
    # 辞書に追加 (値は1-ベースのインデックス i)
    result_dict[prefix] = i

print(result_dict)
```
